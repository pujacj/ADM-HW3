```python
#parse the html page by finding all the anchor tags to get the URLs and store them in a list
links = []
with open("urls.html", "r") as f:  
    contents = f.read()
    soup = BeautifulSoup(contents, 'lxml')  
    for a_tag in soup.find_all("a"):
        links.append(a_tag.text)
```

```python
#create a list to store all the indices of the URLs which throw error
index_lst = []

#iterate over the dataframe of links and crawl data for each index by using time function between each request made to the Wikipedia page
for i,j in df_links["url"].items():
    time.sleep(5)
    try:
        req = requests.get(j)
        with open("articles/article_{}.html".format(i),"w", encoding="utf-8") as file:
            file.write(str(req.content))
        print(i,j,"downloaded")
    except:
        print(i,j,"error")
        index_lst.append(i)
        time.sleep(20*60)
        i = i+1
        continue  
```

```python
title_data = []
intro_data = []
plot_data = []

for i in range(1,10001):
    print(i)
    intro = ""
    plot = ""
    p1,p2,p3,p4,p5,p6,p7,p8 = "","","","","","","",""
    with open("articles/article_{}.html".format(i), "r") as f:  
        contents = f.read()
        soup = BeautifulSoup(contents, 'lxml') 
        #find the title
        title = soup.find("title")
        title_data.append(title.string)
        
        #find the intro (check for maximum of 4 paragraphs under the intro section)
        entrysoup = soup.find('div',attrs={"class":"mw-parser-output"})
        try:
            for strong_tag in entrysoup.findAll('h2'):
                if (strong_tag.get_text() == "Plot[edit]") | (strong_tag.get_text() == "Plot summary[edit]"):
                    intro_para = strong_tag.find_previous_sibling('p')  
                    try:
                        p1 = intro_para.text
                    except:
                        pass
                    try:
                        para_2 = p1.find_previous_sibling('p')
                    except:
                        pass
                    if not para_2:
                        pass
                    else:
                        p2 = para_2.text
                    try:
                        para_3 = para_2.find_previous_sibling('p')
                    except:
                        pass
                    if not para_3:
                        pass
                    else:
                        p3 = para_3.text
                    try:
                        para_4 = para_3.find_previous_sibling('p')
                    except:
                        pass
                    if not para_4:
                        pass
                    else:
                        p4 = para_4.text
        except:
            pass
                                                                              
        #find the plot (check for maximum of 4 paragraphs under the plot section)
        try:
            for strong_tag in entrysoup.findAll('h2'):
                if (strong_tag.get_text() == "Plot[edit]") | (strong_tag.get_text() == "Plot summary[edit]"):
                    plot_para = strong_tag.find_next_sibling('p')
                    try:
                        p5 = plot_para.text
                    except:
                        pass
                    try:
                        para_6 = p5.find_next_sibling('p')
                    except:
                        pass
                    if not para_6:
                        pass
                    else:
                        p6 = para_6.text
                    try:
                        para_7 = para_6.find_next_sibling('p')
                    except:
                        pass
                    if not para_7:
                        pass
                    else:
                        p7 = para_7.text
                    try:
                        para_8 = para_7.find_next_sibling('p')
                    except:
                        pass
                    if not para_8:
                        pass
                    else:
                        p8 = para_8.text
        except:
            pass
    intro = ', '.join(filter(None, (p4,p3,p2,p1)))
    intro_data.append(intro)
    plot = ', '.join(filter(None, (p5,p6,p7,p8)))
    plot_data.append(plot)
```

<h1> Pre Processing </h1>

```python
#convert all the dataframe column data to lowercase for better processing of data
df['Title'] = df['Title'].str.lower()
df['Intro'] = df['Intro'].str.lower()
df['Plot'] = df['Plot'].str.lower()
```

```python
#download the required libraries from nltk
nltk.download('punkt')
nltk.download('stopwords')
stops = stopwords.words("english")
nltk.download('wordnet')
```

```python
print(stops)
```

<h2> Tokenization </h2>

```python
def identify_tokens(row):
    tokens = nltk.word_tokenize(row)
    token_words = [w for w in tokens if w.isalnum()]
    return token_words

#tokenize the Intro and plot columns
df['Title_tokenized'] = df['Title'].apply(identify_tokens)
df['Intro_tokenized'] = df['Intro'].apply(identify_tokens)
df['Plot_tokenized'] = df['Plot'].apply(identify_tokens)
```

```python
df.head(10)
```

<h2> Lemmatizing </h2>

```python
lemmatizer = WordNetLemmatizer()
def lemmatize(row):
    lemmatize_list = [lemmatizer.lemmatize(word) for word in row]
    return (lemmatize_list)

#tokenize the Intro and plot columns
df['Title_lemmatize'] = df['Title_tokenized'].apply(lemmatize)
df['Intro_lemmatize'] = df['Intro_tokenized'].apply(lemmatize)
df['Plot_lemmatize'] = df['Plot_tokenized'].apply(lemmatize)
```

```python
df.head(10)
```

<h2> Removal of stop words </h2>

```python
def remove_stops(row):
    meaningful_words = [w for w in row if not w in stops]
    return (meaningful_words)

#Remove stopwords from the Title, Intro and plot columns
df['Title_sw'] = df['Title_lemmatize'].apply(remove_stops)
df['Intro_sw'] = df['Intro_lemmatize'].apply(remove_stops)
df['Plot_sw'] = df['Plot_lemmatize'].apply(remove_stops)
```

```python
df.head(10)
```

<h2> Removal of punctuation </h2>

```python
symbols = "!\"#$%&()*+-./:;<=>?@[\]^_`{|}~\n"
def remove_punc(row):
    for i in symbols:
        meaningful_words = np.char.replace(str(row), i, ' ')
    return (meaningful_words)

#Remove punctuation from the Title, Intro and plot columns
df['Title_punc'] = df['Title_sw'].apply(remove_punc)
df['Intro_punc'] = df['Intro_sw'].apply(remove_punc)
df['Plot_punc'] = df['Plot_sw'].apply(remove_punc)
```
